```python
def Q6_function(df):
    
    """
    :type : DataFrame
    :rtype: DataFrame
    """
    # TYPE YOUR CODE HERE
    df['Q.4/6.00']=df['Q.4/6.00'].replace(r'-', np.nan)
    df['Q.5/12.00']=df['Q.5/12.00'].replace(r'-', np.nan)
    df['Q.6/6.00']=df['Q.6/6.00'].replace(r'-', np.nan)
    c1=df["Q.4/6.00"].isna().sum()
    c2=df["Q.5/12.00"].isna().sum()
    c3=df["Q.6/6.00"].isna().sum()
    
    df['Q.4/6.00']=df['Q.4/6.00'].fillna(1000)
    df['Q.5/12.00']=df['Q.5/12.00'].fillna(1000)
    df['Q.6/6.00']=df['Q.6/6.00'].fillna(1000)
    
    df["Q.4/6.00"]=df["Q.4/6.00"].astype(float)
    df["Q.4/6.00"]=df["Q.4/6.00"].astype(int)
    df["Q.5/12.00"]=df["Q.5/12.00"].astype(float)
    df["Q.5/12.00"]=df["Q.5/12.00"].astype(int)
    df["Q.6/6.00"]=df["Q.6/6.00"].astype(float)
    df["Q.6/6.00"]=df["Q.6/6.00"].astype(int)

    df["Q.4/6.00"]=df["Q.4/6.00"].replace(1000, (df["Q.4/6.00"].sum()-c1*1000)/(df.shape[0]-c1))
    df["Q.6/6.00"]=df["Q.6/6.00"].replace(1000, (df["Q.6/6.00"].sum()-c3*1000)/(df.shape[0]-c3))
    df["Q.5/12.00"]=df["Q.5/12.00"].replace(1000, (df["Q.5/12.00"].sum()-c2*1000)/(df.shape[0]-c2))
    
    df=df.drop_duplicates(subset=['id', 'Grade/45.00', 'Q.1/5.00', 'Q.2/10.00', 'Q.3/6.00', 'Q.4/6.00', 'Q.5/12.00', 'Q.6/6.00'])
    
    t2=[]
    t2=df.mean()
    t2[0]=df.shape[0]
    
    df.loc[df.shape[0]]=t2
    
    return df
    
# Call the function and print the results.
Q6_df = Q6_function(Q5_df)
print(Q6_df)
```
